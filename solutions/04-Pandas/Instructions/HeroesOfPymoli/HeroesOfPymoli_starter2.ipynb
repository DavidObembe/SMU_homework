```python
# Dependencies and Setup
import pandas as pd
#from IPython.core.interactiveshell import InteractiveShell
#InteractiveShell.ast_node_interactivity = "all"

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
totPlayer_df= purchase_data.SN.value_counts()
totPlayer_df
playercount_df= totPlayer_df.count()
playercount_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
ItemPrice_df= purchase_data.iloc[:,4:]
ItemPrice_df
NumbUnique_df= ItemPrice_df['Item Name'].value_counts()
print(NumbUnique_df.count())
averagePrice_df= ItemPrice_df.Price.mean()
print(averagePrice_df)
totalPurchases_df= purchase_data["Purchase ID"].count()
print(totalPurchases_df)
TotalRevenue_df= ItemPrice_df.Price.sum()
print(TotalRevenue_df)
```

```python
TableSummary= {'Numner of Unique items': [NumbUnique_df.count()], 
                'Average Price': [averagePrice_df], 
                'Number of Purchases': [totalPurchases_df],
                'Total Revenue': [TotalRevenue_df]}
TableSummary_df = pd.DataFrame(TableSummary) 
TableSummary_df.style.format({"Average Price": "${:.2f}", 
                          "Total Revenue": "${:.2f}",})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
mask_male = (purchase_data['Gender']== 'Male')
malePurch_df =purchase_data[mask_male]
unique_male= malePurch_df['SN'].nunique(dropna=True)
print(unique_male)
percentMale= unique_male/playercount_df * 100
print(percentMale)
mask_female = (purchase_data['Gender']== 'Female')
female_df =purchase_data[mask_female]
unique_female= female_df['SN'].nunique(dropna=True)
print(unique_female)
percentfemale_df=unique_female/playercount_df * 100
print(percentfemale_df)
mask_undisclosed = (purchase_data['Gender']== 'Other / Non-Disclosed')
undisclosed_df =purchase_data[mask_undisclosed]
unique_undisclosed= undisclosed_df['SN'].nunique()
print(unique_undisclosed)
percentUndisclosed_df=unique_undisclosed/playercount_df * 100
print(percentUndisclosed_df)
```

```python
GenderDemographic= {'Total count': [unique_male, unique_female, unique_undisclosed], 
                    'Percentage of Players': [percentMale, percentfemale_df, percentUndisclosed_df]}
GenderDemographic_df = pd.DataFrame(GenderDemographic, index =['Male', 'Female', 'Other / Non-disclosed'])
GenderDemographic_df
GenderDemographic_df.style.format({"Percentage of Players": "{:.2f}%"})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data.head(5)
```

```python
genderAnalyst = purchase_data.groupby(['Gender'])
purchaseCount= genderAnalyst['SN'].count()
AverPurchPrice= genderAnalyst['Price'].mean()
print(purchaseCount)
print(AverPurchPrice)
totPurchase=  AverPurchPrice*purchaseCount
print(totPurchase)
averPurPerMale= totPurchase.loc['Male']/unique_male
print(averPurPerMale)
averPurPerFemale=totPurchase.loc['Female']/unique_female
print(averPurPerFemale)
averPurPerOther=totPurchase.loc['Other / Non-Disclosed']/unique_undisclosed
print(averPurPerOther)
```

```python
gender_data= {'Purchase Count': purchaseCount,
                'Average Purchase Price':AverPurchPrice,
                'Total Purchase Value':totPurchase, 
                'Average Purchase Total per Person by Gender':[averPurPerFemale, averPurPerMale, averPurPerOther]}
gender_df= pd.DataFrame(gender_data)
gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Establish bins for ages
purchase_ages= purchase_data
ageBin = [0,10,14,19,24,29,34,39,45,49]
purchase_ages['Age group']= pd.cut(purchase_ages['Age'],ageBin)
purchase_ages.head(5)

#purchase_data
```

```python
#Purchase Count
ageCount= purchase_ages['Age group'].value_counts()
print(ageCount)
```

```python
#Average Purchase Price
#purchase_ages.groupby('binned').mean()
purchasePrice_age = purchase_ages.groupby('Age group').mean()
purchasePrice_age.Price
```

```python
#Total Purchase Value
totalP_byAge = purchase_ages.groupby('Age group').sum()
totalP_byAge.Price
```

```python
#Average Purchase Total per Person by Age Group
avPurperperson= totalP_byAge.Price/ageCount
avPurperperson
```

```python
#Summary Table
age_data= {'Purchase Count': ageCount,
                'Average Purchase Price':purchasePrice_age.Price,
                'Total Purchase Value':totalP_byAge.Price, 
                'Average Purchase Total per Person':avPurperperson}
age_df= pd.DataFrame(age_data)
age_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popularItems= purchase_data.groupby(['Item Name']).count()['Item ID']
popularItems= popularItems.reset_index()
popularItems=popularItems.rename(columns = {'Item ID':'Purchase count'})
popularItems= popularItems.sort_values(['Purchase count'], ascending= False)
popularItems
```

```python
ItemNamePriceID= purchase_data[['Item Name','Price','Item ID']]
ItemNamePriceID=ItemNamePriceID.groupby(['Item Name']).first()
merged_item= popularItems.merge(ItemNamePriceID, on='Item Name', how='inner').iloc[0:5]

merged_item
```

```python
itemTotVal= merged_item['Purchase count']*merged_item['Price']
itemTotVal
merged_item['Total value']= itemTotVal
merged_item=merged_item[['Item Name','Item ID','Price','Purchase count','Total value']]
merged_item
```
