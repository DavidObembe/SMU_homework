```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_students =school_data_complete['student_name'].count()
total_students

#Calculate the total number of students
total_schools = school_data_complete['school_name'].nunique()
total_schools

#Calculate the total budget
total_budget= school_data_complete['budget'].unique().sum()
total_budget

#Calculate the average math score
average_math_score=school_data_complete['math_score'].mean()

#Calculate the average reading score
average_reading_score= school_data_complete['reading_score'].mean()
average_reading_score

#Calculate the percentage of students with a passing math score (70 or greater)
passedMath_mask = (school_data_complete['math_score']>= 70)
passedMath = school_data_complete[passedMath_mask].count().math_score
percent_pmath= (passedMath/total_students)*100

percent_pmath
#Calculate the percentage of students with a passing reading score (70 or greater)
passedreading_mask = (school_data_complete['reading_score']>= 70)
passedReading = school_data_complete[passedreading_mask].count().reading_score
percent_pReading= (passedReading/total_students)*100

percent_pReading

#Calculate the percentage of students who passed math and reading (% Overall Passing)
passedboth_mask = (school_data_complete['reading_score']>= 70)&  (school_data_complete['math_score']>= 70)
passedboth = school_data_complete[passedboth_mask].count().student_name
passedboth
percent_pBoth= (passedboth/total_students)*100
percent_pBoth

#Create a dataframe to hold the above results
data_DistrictSummary= {'Total number of Students': [total_students],
       'Total number of Schools': [total_schools],
      'Total Budget':[total_budget],
      'Average Math Score':[average_math_score],
      'Average Reading Score':[average_reading_score],
      '% Passing Math':[percent_pmath],
       '% Passing Reading':[percent_pReading],
       '% Overall Passing':[percent_pBoth]
      }

df_DistrictSummary = pd.DataFrame(data_DistrictSummary) 
df_DistrictSummary.round(2)
```

## School Summary

* Create a dataframe to hold the above results

```python
school_names= school_data_complete.groupby(['school_name']).count().reset_index()['school_name']
school_names

school_type = school_data.set_index(["school_name"])["type"]
school_type

total_students_per_school= school_data_complete.groupby(['school_name']).count()['size']
total_students_per_school

school_budget= school_data_complete.groupby('school_name').mean()['budget']
school_budget

Per_student_budget = school_budget/total_students_per_school
Per_student_budget

math_mean_per_school= school_data_complete.groupby(['school_name']).mean()['math_score']
math_mean_per_school

reading_mean_per_school= school_data_complete.groupby(['school_name']).mean()['reading_score']
reading_mean_per_school

passed_math_per_school= school_data_complete[passedMath_mask].groupby(['school_name']).count().math_score
percent_Passing_Math = (passed_math_per_school/total_students_per_school)*100
percent_Passing_Math

passed_reading_per_school= school_data_complete[passedreading_mask].groupby(['school_name']).count().reading_score
percent_Passing_reading= (passed_reading_per_school/total_students_per_school)*100
percent_Passing_reading

passed_both_per_school = school_data_complete[passedboth_mask].groupby(['school_name']).count()['reading_score']
percent_Passing_both=(passed_both_per_school/total_students_per_school)*100
percent_Passing_both
```

```python
data_SchoolSummary= {'School Type': school_type,
       'Total Students':total_students_per_school,
      'Total Budget':school_budget,
      'Per Student Budget':Per_student_budget,
      'Average Math Score':math_mean_per_school,
      'Average Reading Score':reading_mean_per_school,
      '% Passing Math':percent_Passing_Math,
       '% Passing Reading':percent_Passing_reading,
       '% Overall Passing':percent_Passing_both
      }

df_SchoolSummary = pd.DataFrame(data_SchoolSummary) 
df_SchoolSummary["Total Budget"] = df_SchoolSummary["Total Budget"].map("${:,.2f}".format)
df_SchoolSummary["Per Student Budget"] = df_SchoolSummary["Per Student Budget"].map("${:,.2f}".format)
#format_dict = {'Total Budget':'${0:,.0f}','Per Student Budget':'${0:,.0f}'}
#df_SchoolSummary= df_SchoolSummary.style.format(format_dict)
#.hide_index()

df_SchoolSummary
```
